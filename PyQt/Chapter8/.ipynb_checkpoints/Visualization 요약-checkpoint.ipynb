```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_ = np.cos(x)

plt.plot(x, y)
plt.plot(x, y_, '-o')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_ = np.cos(x)

plt.plot(x, y, label='sin')
plt.plot(x, y_, '-o', label='cos')

plt.legend()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_ = np.cos(x)

plt.plot(x, y, label='sin')
plt.plot(x, y_, '-o', label='cos')

plt.legend(loc=1) #숫자를 변경하면서 위치 변경됨. (1, 2, 3, 4 중)
```

# scatter

```python
x = np.linspace(0, 10, 20)
y = x ** 2

plt.scatter(x, y, c='r', alpha=0.5) # 여기서 c는 color를 나타냄. color라고 작성해도 무방
plt.show()
```

# histogram

```python
x = [np.random.randint(1, 7) for i in range(100000)]
```

```python
plt.hist(x, bins=11)
plt.show()
```

# Pie Chart

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels)
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels, shadow=True) #shadow : 그림자
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [50, 20, 30]

plt.pie(size, labels=labels, shadow=True, autopct = '%1.2f%%') #shadow : 그림자, autopct : 그래프 내 값 출력
plt.show()
```

# Bar Chart

```python
plt.bar(['one', 'two', 'three'], [10, 20, 30]) # bar : 가로
plt.show()
```

```python
plt.barh(['one', 'two', 'three'], [10, 20, 30]) # barh : 세로
plt.show()
```

# Plotly
 * 공식 홈페이지(https://plotly.com/python/)튜토리얼

```python
import plotly.express as px

x_ = np.array([1, 2, 3, 4, 5])
y_ = x_ ** 2
fig = px.line(x=x_, y=y_)
fig.show()
```

```python
import numpy as np
import plotly.figure_factory as ff

x1,y1 = np.meshgrid(np.arange(0, 2, .2), np.arange(0, 2, .2))
u1 = np.cos(x1)*y1
v1 = np.sin(x1)*y1

fig = ff.create_quiver(x1, y1, u1, v1)

fig.show()
```
